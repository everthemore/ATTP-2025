```python
import numpy as np
```

```python
class BananaExperiment:
    def __init__(self):
        # Nothing to do here
        return

    def reset(self):
         # Randomly choose if the light is on or off
        self.state = np.random.randint(0,2)

        # What we can observe here is the full state
        observation = self.state
        return observation

    def run(self, action):
        # The monkey has the latest observation, and chose and action based on it
        # Now we define what happens based on it:
        reward = 0 # The default is 0, and ...
        if self.state == 1 and action == 1:
            # ... only if light is on and monkey pulls lever, we get a reward
            reward = 1

        # After the experiment, we (choose) to always turn the light off. The monkey sees this
        self.state = 0
        observation = self.state
        
        return observation, reward
```

```python
experiment = BananaExperiment()
```

```python
Q = np.zeros((2,2))
alpha = 0.1
epsilon = 0.1

# Let's run the experiment 10 times
for i in range(1000):

    # First we reset it, and track the observation
    observation = experiment.reset()

    # Now we should make a choice based on the observation! 
    if( np.random.rand() < epsilon ):
        action = np.random.randint(0,2)
    else:
        # Choose action according to Q:
        action = np.argmax( Q[observation, :] )

    # Perform action and get reward
    new_observation, reward = experiment.run(action)

    # Update rule!
    Q[observation, action] = (1-alpha) * Q[observation, action] + alpha*reward
    
    print(f"We observed {observation}, choose to do action {action} and got reward {reward}")

    if observation == 1 and action != 1:
        print("We missed a banana!")
```
